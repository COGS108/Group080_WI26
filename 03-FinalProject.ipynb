## Overview

The rise of AI-generated answers at the top of Google search results — known as AI Overviews — has raised concerns about declining organic traffic to information websites. We investigated whether the U.S. rollout of AI Overviews in May 2024 led to a measurable decline in search-engine-originated clicks to English Wikipedia for stable, evergreen informational topics. Using Wikipedia Clickstream data (monthly counts of search-to-article transitions) and Google Trends (relative search interest), we compared the pre-rollout period (June 2023) to the post-rollout period (June 2024), controlling for seasonality by matching calendar months and restricting to topics whose search interest remained stable.

At the aggregate level, total search-originated clicks to Wikipedia declined by approximately 5.4% year-over-year from June 2023 to June 2024. A Wilcoxon signed-rank test on a 200,000-article subsample drawn from 3.4 million matched articles confirmed a statistically significant shift in per-article click distributions (p < 0.001), with a mean decline of approximately 82 clicks per article and a negligible standardized effect size (Cohen's d ≈ -0.009). However, results at the topic level were mixed: the normalized clicks-per-interest ratio did not decline for Climate change and was unstable for Quantum mechanics because Trends interest was near zero. January 2025 also showed a rebound in total clicks, complicating a simple monotonic-decline narrative. These findings are directionally consistent with industry reports of reduced organic click-through rates, but they suggest substantial heterogeneity by topic and time window rather than a uniform decline.

## Background and Prior Work

The way we search for information online is going through a huge change right now. For about 30 years, Google and other search engines worked pretty simply: you typed in a question, and they gave you a list of websites to click on. This "Ten Blue Links" model meant that users had to do the work of clicking through different sites, reading them, and figuring out what was true on their own.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) But now, with AI being added to search engines, things are different. Instead of just showing you where to find answers, search engines are now trying to give you the answer directly. Researchers call this shift moving from "Information Retrieval" to "Generative Information Retrieval."

This change didn't happen overnight — it built up over several years. In 2015, Google started using something called RankBrain, which was the first time they used deep learning to understand search queries better. Then in 2019, they added BERT (Bidirectional Encoder Representations from Transformers), which helped Google understand words in the context of a full sentence instead of just individually.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) In 2021, they introduced MUM (Multitask Unified Model), and finally in 2023, they launched the Search Generative Experience (SGE), which became "AI Overviews" in 2024. This was the first time Google actually showed AI-generated text at the top of search results instead of just links.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

One major consequence of this shift is what researchers call the "Great Decoupling" — more people are searching than ever, but fewer people are actually clicking through to websites. The numbers are striking: about 60% of Google searches now end without the user clicking on anything (rising to 77% on mobile).<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) A study by Seer Interactive found that when an AI Overview shows up, click-through rates for the top search results dropped from 1.41% to just 0.64% — more than a 50% decrease.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) Another study from Ahrefs found that AI Overviews reduce organic clicks by around 58%, and the trend is worsening over time.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)

Beyond the economic impact on websites, there are also concerns about whether these AI-generated answers are actually reliable. When you see a list of links, you know you need to evaluate them yourself. But when an AI gives you a direct answer, it can seem more trustworthy than it actually is — researchers call this the "Illusion of Neutrality."<a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7) AI systems can "hallucinate," sometimes making up information that sounds believable but is completely wrong. There have been cases where AI summaries cited satirical articles as if they were real news, or fabricated citations that don't exist.<a name="cite_ref-8"></a>[<sup>8</sup>](#cite_note-8) Research from Stanford has also found that AI models like Gemini can have political biases built into them based on training data and fine-tuning.<a name="cite_ref-9"></a>[<sup>9</sup>](#cite_note-9)

Some researchers have started creating tools to evaluate these AI search systems. The C-SEO Bench dataset benchmarks how different writing styles affect whether content appears in AI-generated answers.<a name="cite_ref-10"></a>[<sup>10</sup>](#cite_note-10) Salesforce created the Answer Engine Eval framework, which measures how well AI answers cover a topic and whether cited sources are reliable.<a name="cite_ref-11"></a>[<sup>11</sup>](#cite_note-11) However, there are still few long-term public-data studies that directly compare pre/post behavioral shifts after AI Overviews launched — this is the empirical gap our research addresses.

A final validity point is the distinction between **industry reporting** and **causal evidence**. Industry studies provide useful directional signals, but many are based on proprietary methods and cannot be fully audited. Our approach emphasizes reproducibility with public datasets, explicit assumptions, and transparent statistical testing, while treating all conclusions as associational rather than causal.

## Hypothesis


**H₀ (Null Hypothesis):** The rollout of Google AI Overviews in May 2024 had no effect on the rate of search-engine-originated clicks to Wikipedia for stable, evergreen informational topics. The distribution of per-article click counts and the normalized clicks-per-interest ratio are the same in the pre-rollout period (June 2023) and the post-rollout period (June 2024).

**H₁ (Alternative Hypothesis):** The rollout of Google AI Overviews led to a statistically significant decline in search-engine-originated clicks to Wikipedia for stable, evergreen informational topics. The per-article click counts and the normalized clicks-per-interest ratio are lower in the post-rollout period than in the pre-rollout period.

We expect this effect specifically for informational queries — topics where users are seeking factual answers that AI Overviews can directly provide — rather than navigational queries (where users want a specific website) or transactional queries (where users want to complete an action). We also hypothesize that adoption may be nonlinear: modest initial changes followed by a steeper decline as users become accustomed to consuming AI-generated summaries without clicking through.

**Primary outcome variable:** Monthly Wikipedia Clickstream count of transitions where `prev = 'other-search'` (search-engine-originated clicks), filtered to evergreen informational articles.

**Control variables:** Google Trends interest stability (topics must satisfy |post median − pre median| ≤ 5 Trends points); seasonality controlled by comparing same calendar month (June-to-June).

**Significance level:** α = 0.05, two-tailed for the formal tests.


## Dataset(s)

### Data overview

We use two publicly available datasets to proxy zero-click search behavior:

- **Dataset #1 — Wikipedia Clickstream** (clicks from search engines to Wikipedia articles)
  - **Source:** Wikimedia Clickstream (English Wikipedia, monthly dumps)
  - **Link:** https://dumps.wikimedia.org/other/clickstream/
  - **Observations:** One row per (referrer, destination article) pair per monthly file; millions of pairs per file.
  - **Variables:** `prev` (referrer — e.g. `other-search` for external search engines, or an article title), `curr` (destination Wikipedia article), `type` (link/external/other), `n` (count of that transition in the month).
  - **Months available:** June 2023 (pre-AIO), June 2024 (post-AIO), January 2025 (post-AIO).
  - **Relevance:** Filtering `prev = 'other-search'` gives monthly counts of clicks from search engines to Wikipedia. A decline in these counts after mid-2024 for informational articles is consistent with increased zero-click behavior.
  - **Shortcomings:** Aggregated — cannot distinguish Google from other search engines. Only captures traffic to Wikipedia. Pairs with ≤10 observations are excluded at source.

- **Dataset #2 — Google Trends** (relative search interest over time)
  - **Source:** Google Trends (via `pytrends` Python library)
  - **Link:** https://trends.google.com/
  - **Observations:** One row per week per keyword.
  - **Variables:** `date` (week start), `keyword` (search term), `interest` (0–100 relative popularity index).
  - **Time range:** January 2023 – February 2025 (weekly).
  - **Relevance:** Shows whether people kept searching for topics at similar rates. Stable or rising Trends interest with falling clickstream counts supports that answers are being consumed in-search.
  - **Shortcomings:** Relative scale (0–100), not absolute volume. Sampling/API rate limits. Geographic and category choices affect results.

**Combining the datasets:** We align both datasets by **month** and **topic**. For each evergreen informational topic, we compute monthly Wikipedia clickstream counts from search to the matching article, get monthly Google Trends interest for the corresponding keyword, and compare pre- vs post-AI Overviews periods. We compute a normalized metric — **clicks per unit of Trends interest** — to control for underlying search demand.

| Variable | Source | Type | Description |
|----------|--------|------|-------------|
| `prev` | Clickstream | Categorical | Referrer (e.g. `other-search`) |
| `curr` | Clickstream | Categorical | Destination Wikipedia article title |
| `n` | Clickstream | Numeric (int) | Count of that transition in the month |
| `month` | Both | Date (YYYY-MM) | Calendar month |
| `keyword` | Google Trends | Categorical | Search term queried |
| `interest` | Google Trends | Numeric (0–100) | Relative search popularity |
| `clicks_per_interest` | Derived | Numeric (float) | `clicks_from_search / interest` — normalized metric |
| `period` | Derived | Binary | Pre-AIO (before May 2024) vs Post-AIO |

```python
# --- Core cleaning for Clickstream ---
cs_by_article = cs_by_article.copy()
cs_monthly = cs_monthly.copy()

# Normalize key columns
cs_by_article['month'] = cs_by_article['month'].astype(str).str.strip()
cs_by_article['curr'] = cs_by_article['curr'].astype(str).str.strip()
cs_by_article['clicks_from_search'] = pd.to_numeric(cs_by_article['clicks_from_search'], errors='coerce')

cs_monthly['month'] = cs_monthly['month'].astype(str).str.strip()
cs_monthly['clicks_from_search'] = pd.to_numeric(cs_monthly['clicks_from_search'], errors='coerce')

# Resolve any accidental duplicate keys at (month, article)
dup_pairs = cs_by_article.duplicated(subset=['month', 'curr']).sum()
if dup_pairs > 0:
    cs_by_article = (
        cs_by_article
        .groupby(['month', 'curr'], as_index=False)['clicks_from_search']
        .sum()
    )

# Keep only valid rows used in this report
cs_by_article = cs_by_article.dropna(subset=['month', 'curr', 'clicks_from_search'])
cs_monthly = cs_monthly.dropna(subset=['month', 'clicks_from_search'])
cs_by_article = cs_by_article[cs_by_article['clicks_from_search'] >= 0]
cs_monthly = cs_monthly[cs_monthly['clicks_from_search'] >= 0]

# --- Core cleaning for Google Trends ---
trends_long = trends_long.copy()
trends_long['keyword'] = trends_long['keyword'].astype(str).str.strip()
trends_long['interest'] = pd.to_numeric(trends_long['interest'], errors='coerce')
trends_long['date'] = pd.to_datetime(trends_long['date'], errors='coerce')
trends_long = trends_long.dropna(subset=['keyword', 'interest', 'date'])
trends_long = trends_long[trends_long['interest'] >= 0]
trends_long['month'] = trends_long['date'].dt.to_period('M').astype(str)

# Recreate period labels after cleaning
cs_monthly['period'] = np.where(cs_monthly['month'] < '2024-05', 'Pre-AIO', 'Post-AIO')
trends_long['period'] = np.where(trends_long['date'] < '2024-05-14', 'Pre-AIO', 'Post-AIO')

print('=== Post-cleaning quality checks ===')
print(f'Clickstream duplicate (month, curr) pairs resolved: {dup_pairs}')
print(f'Clickstream rows: {len(cs_by_article):,}')
print(f'Monthly totals rows: {len(cs_monthly):,}')
print(f'Trends rows: {len(trends_long):,}')
print(f'Any negative click counts: {(cs_by_article["clicks_from_search"] < 0).any()}')
print(f'Any missing in clickstream core fields: {cs_by_article[["month", "curr", "clicks_from_search"]].isnull().any().any()}')
print(f'Any missing in trends core fields: {trends_long[["date", "keyword", "interest", "month"]].isnull().any().any()}')
```

## Data Analysis & Results

### Exploratory Data Analysis

Our analysis proceeds in three stages:

1. **Aggregate Click Trends (Descriptive Context)** — We examine total and per-article search-originated clicks across our three monthly snapshots (June 2023, June 2024, January 2025) to establish the broad landscape.
2. **Topic-Level Clicks-Per-Interest Analysis** — We restrict to evergreen topics with stable Google Trends interest and compute a normalized metric (clicks per unit of interest) for pre/post comparison.
3. **Statistical Inference** — We apply formal statistical tests (Wilcoxon signed-rank, permutation test, bootstrapped confidence intervals) to quantify the significance and magnitude of any observed change.

The EDA below is the final subset focused on the variables and comparisons needed to answer the research question.

### Topic-Level Combined Analysis: Clicks per Unit of Interest

This section implements the operationalized test aligned to our research question. We:

1. **Restrict** the clickstream data to evergreen informational articles that passed the Trends stability check above.
2. **Merge** clickstream counts with monthly Google Trends interest at the topic level.
3. **Compute** a normalized metric — clicks per unit of Trends interest — which controls for changes in underlying search demand.

If AI Overviews are intercepting traffic, we expect this ratio to decline even when search interest remains stable.

- **Figure 4 takeaway:** Topic interest is mostly stable for Climate change, while Quantum mechanics remains near zero for much of the timeline.
- **Figure 5 takeaway:** The normalized metric does not show a clean post-rollout decline for the available topics.
- **Figure 6 takeaway:** Topic trajectories are heterogeneous, so aggregate conclusions may mask topic-specific behavior.

```python
# ===== Build topic-level merged dataset =====
article_for_keyword = {k: v for k, v in evergreen_topics.items()}
stable_articles = [article_for_keyword[k] for k in stable_keywords if k in article_for_keyword]

cs_topic = cs_by_article[cs_by_article['curr'].isin(stable_articles)].copy()
reverse_map = {v: k for k, v in article_for_keyword.items()}
cs_topic['keyword'] = cs_topic['curr'].map(reverse_map)

trends_topic = trends_long[trends_long['keyword'].isin(stable_keywords)].copy()
trends_topic_monthly = trends_topic.groupby(['month', 'keyword'], as_index=False)['interest'].mean()

combined_topic = cs_topic.merge(
    trends_topic_monthly,
    on=['month', 'keyword'],
    how='inner'
)
combined_topic['clicks_per_interest'] = (
    combined_topic['clicks_from_search'] / combined_topic['interest'].clip(lower=0.1)
)
combined_topic['period'] = np.where(combined_topic['month'] < '2024-05', 'Pre-AIO', 'Post-AIO')

combined_monthly = combined_topic.groupby('month', as_index=False).agg(
    wiki_clicks=('clicks_from_search', 'sum'),
    interest=('interest', 'mean'),
    clicks_per_interest=('clicks_per_interest', 'mean')
)
combined_monthly['period'] = np.where(combined_monthly['month'] < '2024-05', 'Pre-AIO', 'Post-AIO')

print("Combined topic-level dataset (stable evergreen topics only):")
display(combined_topic)
print("\nMonthly aggregate:")
display(combined_monthly)

# ===== Figure 4: Google Trends interest over time =====
fig, ax = plt.subplots(figsize=(12, 5))
palette = {'Quantum mechanics': '#2196F3', 'Climate change': '#FF5722'}

for kw in trends_long['keyword'].unique():
    kw_data = trends_long[trends_long['keyword'] == kw].sort_values('date')
    ax.plot(kw_data['date'], kw_data['interest'], label=kw,
            color=palette.get(kw, '#666'), linewidth=1.5, alpha=0.85)

ax.axvline(pd.Timestamp('2024-05-14'), color='red', linestyle='--', alpha=0.7,
           label='AI Overviews rollout (May 14, 2024)')
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Google Trends Interest (0–100)', fontsize=12)
ax.set_title('Google Trends Interest for Evergreen Topics Over Time', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('results/fig4_trends_stability.png', dpi=150, bbox_inches='tight')
plt.show()

# ===== Figure 5: Combined panel — clicks, interest, and ratio =====
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
period_colors = {'Pre-AIO': '#2196F3', 'Post-AIO': '#FF5722'}
bar_colors = [period_colors[p] for p in combined_monthly['period']]

axes[0].bar(combined_monthly['month'], combined_monthly['wiki_clicks'] / 1e3,
            color=bar_colors, edgecolor='white')
axes[0].set_ylabel('Clicks (Thousands)')
axes[0].set_title('Evergreen Topic Clicks from Search')
axes[0].tick_params(axis='x', rotation=45)

axes[1].bar(combined_monthly['month'], combined_monthly['interest'],
            color=bar_colors, edgecolor='white')
axes[1].set_ylabel('Avg Trends Interest (0–100)')
axes[1].set_title('Evergreen Topic Search Interest')
axes[1].tick_params(axis='x', rotation=45)

axes[2].bar(combined_monthly['month'], combined_monthly['clicks_per_interest'] / 1e3,
            color=bar_colors, edgecolor='white')
axes[2].set_ylabel('Clicks per Interest Unit (Thousands)')
axes[2].set_title('Normalized: Clicks / Interest')
axes[2].tick_params(axis='x', rotation=45)

legend_elements = [Patch(facecolor='#2196F3', label='Pre-AIO'),
                   Patch(facecolor='#FF5722', label='Post-AIO')]
fig.legend(handles=legend_elements, loc='upper center', ncol=2, fontsize=11,
           bbox_to_anchor=(0.5, 1.05))
fig.suptitle('Topic-Level Clickstream vs Trends (Stable Evergreen Topics)',
             fontsize=14, fontweight='bold', y=1.1)
plt.tight_layout()
plt.savefig('results/fig5_combined_panel.png', dpi=150, bbox_inches='tight')
plt.show()

# ===== Figure 6: Per-topic trajectories =====
fig, ax = plt.subplots(figsize=(10, 5))
stable_article_map = {evergreen_topics[k]: k for k in stable_keywords if k in evergreen_topics}
topic_clicks = cs_by_article[cs_by_article['curr'].isin(stable_article_map.keys())].copy()
topic_clicks['topic'] = topic_clicks['curr'].map(stable_article_map)

for topic in sorted(topic_clicks['topic'].unique()):
    t = topic_clicks[topic_clicks['topic'] == topic].sort_values('month')
    ax.plot(t['month'], t['clicks_from_search'], marker='o', linewidth=2,
            markersize=8, label=topic, color=palette.get(topic, '#666'))
    for _, row in t.iterrows():
        ax.annotate(f"{row['clicks_from_search']:,.0f}", (row['month'], row['clicks_from_search']),
                    textcoords='offset points', xytext=(0, 10), ha='center', fontsize=8)

ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Clicks from Search', fontsize=12)
ax.set_title('Search Clicks to Stable Evergreen Topics Over Time', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('results/fig6_topic_trajectories.png', dpi=150, bbox_inches='tight')
plt.show()

print("Figure 5 shows the operationalized comparison: clicks, interest, and their ratio.")
print("Figure 6 traces per-topic click trajectories across the three available months.")
```

### Statistical Inference: Pre vs Post AI Overviews

We now apply formal statistical tests to quantify whether the observed decline is statistically significant. Our primary comparison is **June 2023 (pre-AIO) vs June 2024 (post-AIO)**, which controls for seasonality by comparing the same calendar month one year apart.

We conduct three analyses:
1. **Wilcoxon signed-rank test** on matched articles (articles present in both months), testing whether per-article clicks systematically decreased.
2. **Permutation test** on the mean clicks-per-interest ratio, testing whether the observed pre/post difference could arise by chance.
3. **Bootstrapped 95% confidence interval** for the mean difference in per-article clicks between periods.

All tests use α = 0.05 and we report effect sizes (Cohen's d) alongside p-values to assess practical significance.

Interpretation rule used in this notebook:
- **Statistical significance** asks whether a change is unlikely under the null.
- **Practical significance** asks whether the magnitude is meaningful in real terms.

Because the click distribution is extremely skewed and sample sizes are very large, small effects can still produce very small p-values.

```python
print("=" * 70)
print("STATISTICAL ANALYSIS: June 2023 (Pre-AIO) vs June 2024 (Post-AIO)")
print("=" * 70)

# --- 1. Wilcoxon signed-rank test on matched articles ---
pre = cs_by_article[cs_by_article['month'] == '2023-06'][['curr', 'clicks_from_search']].copy()
post = cs_by_article[cs_by_article['month'] == '2024-06'][['curr', 'clicks_from_search']].copy()

matched = pre.merge(post, on='curr', suffixes=('_pre', '_post'))
matched['diff'] = matched['clicks_from_search_post'] - matched['clicks_from_search_pre']

print(f"\n--- Test 1: Wilcoxon Signed-Rank Test (matched articles) ---")
print(f"Articles in both months: {len(matched):,}")
print(f"Mean pre clicks:  {matched['clicks_from_search_pre'].mean():,.1f}")
print(f"Mean post clicks: {matched['clicks_from_search_post'].mean():,.1f}")
print(f"Mean difference:  {matched['diff'].mean():,.1f}")
print(f"Median difference: {matched['diff'].median():,.1f}")

# Subsample for Wilcoxon (3.4M pairs is computationally intensive);
# we sample 200,000 matched pairs for a tractable but still very powerful test.
rng = np.random.default_rng(42)
sample_size = min(200_000, len(matched))
sample_idx = rng.choice(len(matched), size=sample_size, replace=False)
sample = matched.iloc[sample_idx]

wilcoxon_stat, wilcoxon_p = stats.wilcoxon(
    sample['clicks_from_search_pre'],
    sample['clicks_from_search_post'],
    alternative='two-sided'
)
print(f"\nWilcoxon statistic (n={sample_size:,} subsample): {wilcoxon_stat:,.0f}")
print(f"p-value: {wilcoxon_p:.2e}")
print(f"Significant at α=0.05: {wilcoxon_p < 0.05}")

pooled_std = np.sqrt(
    (matched['clicks_from_search_pre'].std()**2 + matched['clicks_from_search_post'].std()**2) / 2
)
cohens_d = matched['diff'].mean() / pooled_std
print(f"Cohen's d (effect size): {cohens_d:.4f}")
print(f"  Interpretation: {'negligible' if abs(cohens_d) < 0.2 else 'small' if abs(cohens_d) < 0.5 else 'medium' if abs(cohens_d) < 0.8 else 'large'} effect")

# --- 2. Permutation test on clicks-per-interest ratio ---
print(f"\n--- Test 2: Permutation Test (clicks-per-interest ratio) ---")

# Filter to Climate change only (Quantum mechanics has near-zero Trends interest,
# making clicks/interest artificially dominated by the 0.1 floor clipping).
cc_combined = combined_topic[combined_topic['keyword'] == 'Climate change'].copy()
pre_ratio = cc_combined[cc_combined['period'] == 'Pre-AIO']['clicks_per_interest'].values
post_ratio = cc_combined[cc_combined['period'] == 'Post-AIO']['clicks_per_interest'].values

# Also show the full topic-level data for transparency
print("Topic-level clicks-per-interest (all stable evergreen topics):")
display(combined_topic[['month', 'keyword', 'clicks_from_search', 'interest', 'clicks_per_interest', 'period']])
print("\nNote: 'Quantum mechanics' has near-zero Trends interest, making the ratio unreliable.")
print("Permutation test restricted to 'Climate change' for methodological validity.\n")

if len(pre_ratio) > 0 and len(post_ratio) > 0:
    observed_diff = pre_ratio.mean() - post_ratio.mean()
    all_ratios = np.concatenate([pre_ratio, post_ratio])
    n_pre = len(pre_ratio)
    n_permutations = 10000

    perm_diffs = np.zeros(n_permutations)
    for i in range(n_permutations):
        shuffled = rng.permutation(all_ratios)
        perm_diffs[i] = shuffled[:n_pre].mean() - shuffled[n_pre:].mean()

    perm_p = (np.abs(perm_diffs) >= np.abs(observed_diff)).mean()

    print(f"Observed mean difference (pre - post): {observed_diff:,.1f}")
    print(f"Direction: {'Pre > Post (decline)' if observed_diff > 0 else 'Post > Pre (increase)'}")
    print(f"Permutation p-value (two-sided, {n_permutations:,} permutations): {perm_p:.4f}")
    print(f"Significant at α=0.05: {perm_p < 0.05}")

    fig, ax = plt.subplots(figsize=(8, 4))
    ax.hist(perm_diffs, bins=50, color='#90CAF9', edgecolor='white', alpha=0.8,
            label='Permutation distribution')
    ax.axvline(observed_diff, color='red', linewidth=2, linestyle='--',
               label=f'Observed diff = {observed_diff:,.0f}')
    ax.set_xlabel('Mean Difference (Pre − Post)', fontsize=11)
    ax.set_ylabel('Count', fontsize=11)
    ax.set_title('Permutation Test: Climate Change Clicks-per-Interest',
                 fontsize=13, fontweight='bold')
    ax.legend(fontsize=10)
    plt.tight_layout()
    plt.savefig('results/fig7_permutation_test.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    perm_p = float('nan')
    observed_diff = float('nan')
    print("Insufficient data for permutation test.")

# --- 3. Bootstrap CI for mean difference in per-article clicks ---
print(f"\n--- Test 3: Bootstrapped 95% CI for Mean Difference ---")

n_bootstrap = 10000
boot_diffs = np.zeros(n_bootstrap)
diffs = matched['diff'].values
# Subsample for bootstrap tractability
boot_sample = rng.choice(diffs, size=min(200_000, len(diffs)), replace=False)

for i in range(n_bootstrap):
    resample = rng.choice(boot_sample, size=len(boot_sample), replace=True)
    boot_diffs[i] = resample.mean()

ci_lower, ci_upper = np.percentile(boot_diffs, [2.5, 97.5])
print(f"Mean difference (post - pre): {diffs.mean():,.2f}")
print(f"95% Bootstrap CI: [{ci_lower:,.2f}, {ci_upper:,.2f}]")
ci_excludes_zero = (ci_lower > 0) or (ci_upper < 0)
print(f"CI excludes zero: {ci_excludes_zero}")

# --- Summary table ---
print("\n" + "=" * 70)
print("SUMMARY OF STATISTICAL TESTS")
print("=" * 70)
summary_data = {
    'Test': ['Wilcoxon signed-rank (articles)', 'Permutation (Climate change ratio)', 'Bootstrap 95% CI (articles)'],
    'Statistic / CI': [
        f'W = {wilcoxon_stat:,.0f}',
        f'Obs. diff = {observed_diff:,.0f}',
        f'[{ci_lower:,.1f}, {ci_upper:,.1f}]'
    ],
    'p-value': [f'{wilcoxon_p:.2e}', f'{perm_p:.4f}', 'N/A (CI-based)'],
    'Significant (α=0.05)': [wilcoxon_p < 0.05, perm_p < 0.05 if not np.isnan(perm_p) else 'N/A', ci_excludes_zero],
    'Direction': [
        'Post < Pre (decline)' if matched['diff'].mean() < 0 else 'Post > Pre',
        'Post > Pre (no decline)' if observed_diff < 0 else 'Pre > Post (decline)',
        'Decline' if ci_upper < 0 else ('Increase' if ci_lower > 0 else 'Includes zero')
    ]
}
display(pd.DataFrame(summary_data))

# --- Robustness: include Jan 2025 ---
print("\n--- Robustness Check: Including January 2025 ---")
post_jan = cs_by_article[cs_by_article['month'] == '2025-01'][['curr', 'clicks_from_search']].copy()
matched_jan = pre.merge(post_jan, on='curr', suffixes=('_pre', '_jan25'))
matched_jan['diff'] = matched_jan['clicks_from_search_jan25'] - matched_jan['clicks_from_search_pre']

if len(matched_jan) > 0:
    jan_sample_idx = rng.choice(len(matched_jan), size=min(200_000, len(matched_jan)), replace=False)
    jan_sample = matched_jan.iloc[jan_sample_idx]
    w_stat_jan, w_p_jan = stats.wilcoxon(
        jan_sample['clicks_from_search_pre'],
        jan_sample['clicks_from_search_jan25'],
        alternative='two-sided'
    )
    print(f"Matched articles (June 2023 vs Jan 2025): {len(matched_jan):,}")
    print(f"Mean pre (June 2023): {matched_jan['clicks_from_search_pre'].mean():,.1f}")
    print(f"Mean Jan 2025: {matched_jan['clicks_from_search_jan25'].mean():,.1f}")
    print(f"Mean diff: {matched_jan['diff'].mean():,.1f}")
    print(f"Wilcoxon p-value (n={len(jan_sample):,} subsample): {w_p_jan:.2e}")
    print(f"Significant: {w_p_jan < 0.05}")
    print("Note: June vs January comparison confounds seasonal effects.")
    print("This robustness check examines whether the decline persists or reverses.")
```

## Conclusion & Discussion

### Summary of Findings

Our analysis investigated whether the May 2024 rollout of Google AI Overviews was associated with a decline in search-engine-originated clicks to Wikipedia for stable, evergreen informational topics. Three key results emerged:

1. **Aggregate decline:** Total search-to-Wikipedia clicks declined by approximately 5.4% year-over-year from June 2023 to June 2024, while January 2025 showed a rebound. The June-to-June comparison controls for seasonality and is directionally consistent with AI Overviews intercepting organic traffic.

2. **Article-level significance:** A Wilcoxon signed-rank test on a 200,000-article subsample (from 3.4 million matched articles present in both June 2023 and June 2024) confirmed a statistically significant shift in per-article click distributions (p < 0.001). The full matched set showed a mean decline of approximately 82 clicks per article and a median decline of 4 clicks. Cohen's d was negligible, reflecting the heavily right-skewed distribution and the difference between statistical and practical significance.

3. **Mixed topic-level results:** Our normalized clicks-per-interest ratio — the primary operationalized metric — did **not** show the expected decline for the two evergreen topics tested. For Climate change, the ratio increased slightly from pre to post, and Quantum mechanics had near-zero Google Trends interest (rendering the ratio unreliable). This suggests that well-known, high-authority Wikipedia articles may be partially insulated from the zero-click effect — or may even benefit from AI Overview citations that drive users to verify information at the source.

### Connection to Prior Work

Our aggregate findings are directionally consistent with industry reports from Seer Interactive and Ahrefs, which documented 50–60% reductions in organic click-through rates when AI Overviews appear. However, the magnitude of our observed aggregate decline (~5.4%) is notably smaller, and the topic-level results add important nuance. This discrepancy likely reflects several factors: (a) Wikipedia's unique position as a high-authority reference site that Google frequently cites within AI Overviews, potentially driving *more* clicks to Wikipedia for well-known topics even as overall click-through rates decline; (b) our `other-search` proxy captures all search engines, not just Google, diluting the Google-specific effect; and (c) the industry studies measured click-through rates at the query level with direct experimental controls, while our proxy approach is inherently noisier.

The "Great Decoupling" described in our background — rising search volume paired with declining click-through rates — is partially consistent with what we observe at the aggregate level. However, the topic-level results suggest the effect is heterogeneous: some article categories may experience significant traffic losses while high-authority reference pages on canonical topics may actually benefit. This heterogeneity is an important finding that is not captured by the aggregate industry statistics.

### Contradictions and Complications

The most notable complication is the **January 2025 rebound** in total clicks. This could reflect: (a) seasonal effects (January may have different search patterns than June — e.g., return to school/work driving informational queries); (b) Google adjusting AI Overview coverage or display frequency; (c) growth in Wikipedia's overall article base attracting new traffic; or (d) users becoming more skeptical of AI-generated answers and returning to source verification. Without continuous monthly data, we cannot distinguish these explanations.

### Limitations

Several important limitations constrain the strength of our conclusions:

- **Sparse temporal coverage:** With only three monthly snapshots, we cannot establish a trend line, control for seasonality robustly, or distinguish a permanent shift from a temporary fluctuation. Continuous monthly data across the full 2023–2025 period would substantially strengthen the analysis.

- **Narrow topic set:** Our operationalized test uses only two evergreen topics (Quantum mechanics, Climate change) that could be mapped between Trends keywords and Wikipedia article titles. A larger, more diverse topic set would improve generalizability and statistical power.

- **Proxy limitations:** The `other-search` referrer aggregates all search engines — we cannot isolate Google's contribution. If Bing or DuckDuckGo traffic increased while Google traffic declined, the net effect would be attenuated. Wikipedia clickstream also only captures one slice of the web; effects on smaller sites may be much larger.

- **Relative Trends scale:** Google Trends' 0–100 normalization means we measure relative interest, not absolute search volume. If overall search volume grew substantially, the Trends scale would mask that change.

- **No causal identification:** Our pre/post comparison is observational, not experimental. Many other changes occurred between June 2023 and June 2024 (algorithm updates, cultural shifts, Wikipedia content changes). We can establish association but not causation.

### Future Directions

If this work were to continue beyond this class, the most impactful next steps would be:

1. **Acquire continuous monthly clickstream data** (all available months from 2022–2025) to build a proper interrupted time-series design with seasonal controls.
2. **Expand the evergreen topic universe** to 50+ topics with explicit inclusion/exclusion criteria, enabling more powerful statistical tests.
3. **Add comparison groups** — e.g., navigational topics (like company websites) that should not be affected by AI Overviews — as a control condition.
4. **Supplement with alternative data sources** such as SimilarWeb or Datos clickstream panels to validate findings with different measurement approaches.
5. **Investigate heterogeneity** — do some topic categories (science, history, health) show larger effects than others?

### Conclusion

Our analysis provides preliminary evidence that the introduction of Google AI Overviews is associated with a modest but statistically significant decline in search-engine-originated clicks to Wikipedia for informational topics. This finding aligns with broader industry observations of the zero-click search trend, though the magnitude is smaller than reported in proprietary studies. The January 2025 rebound introduces important ambiguity that warrants further investigation with denser temporal data. Despite its limitations, this study demonstrates that the impact of AI-generated search answers on organic web traffic can be detected using freely available public datasets — an important methodological contribution given that most existing evidence comes from expensive proprietary data sources.
